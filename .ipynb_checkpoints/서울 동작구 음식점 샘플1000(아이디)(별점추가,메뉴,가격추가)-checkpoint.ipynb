```python
# 패키지 임포트
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import time
from tqdm import tqdm_notebook
import urllib
from urllib.request import urlopen
```

```python
df = pd.read_csv('./서울 동작구 음식점 샘플1000(아이디)(별점추가).csv',encoding='cp949')
df
```

```python
df = df[df['지번주소'].str.contains('사당동')==True]
```

```python
df
```

```python
kind = pd.read_csv('./서울시 착한가격음식점(종합).csv',encoding='cp949',index_col=0)
kind = kind[kind['행정동'] == '사당동']
kind.식당아이디
kind.rename(columns={'업소명':'사업장명'},inplace=True)
```

```python
# 착한가격업소 가게와 중복되는 일반음식점 제외
kind.merge(df,how='inner',on='사업장명')
```

```python
kind[kind.사업장명 == '산천칡냉면']
kind[kind.사업장명 == '마포곱창']
kind[kind.사업장명 == '달길포차']
kind[kind.사업장명 == '황제소고기국밥']
kind[kind.사업장명 == '은혜식당']
```

```python
df[df.사업장명 == '산천칡냉면']
df[df.사업장명 == '마포곱창']
df[df.사업장명 == '달길포차']
df[df.사업장명 == '황제소고기국밥']
df[df.사업장명 == '은혜식당']
```

```python
# 산천칡냉면,마포곱창,달길포차,황제소고기국밥,은혜식당 주소동일 중복 데이터 확인
# 제거 후 일반음식점 메뉴가격 크롤링
df.drop([215,649,615,93,142],inplace=True)
df
```

```python
df.reset_index(inplace=True)
df
```

```python
menu_list =[]
price_list = []
for n in tqdm_notebook(range(len(df))) :
    try :
        urlnum = df['업소id'][n]
        url = f'https://pcmap.place.naver.com/restaurant/{urlnum}/menu/list'
        
        driver = webdriver.Chrome('./driver/chromedriver')
        driver.get(url)
        time.sleep(3)
        html = urlopen(driver.current_url)
        bs_obj = BeautifulSoup(html,'html.parser')

        menu = bs_obj.find('span',{'class':'_3yfZ1'}).text
        price = bs_obj.find('div',{'class':'_3qFuX'}).text
        
        menu_list.append(menu)
        price_list.append(price)
        
    except :
        menu_list.append("-")
        price_list.append("-")
```

```python
df['대표메뉴'] = menu_list
df['가격정보'] = price_list
df
```

```python
df.drop(['index'],axis=1,inplace=True)
```

```python
df.to_csv('./서울시 동작구 일반음식점(메뉴가격 추가).csv',encoding='cp949')
```
